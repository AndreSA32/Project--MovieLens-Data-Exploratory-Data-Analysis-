### 8. Visualize how popularity of genres has changed over the years - 10 marks

Note that you need to use the **percent of number of releases in a year** as a parameter of popularity of a genre

Hint 1: You need to reach to a data frame where the release year is the index and the genre is the column names (one cell shows the number of release in a year in one genre) or vice versa. (Drop unnecessary column if there are any)

Hint 2: Find the total number of movies release in a year(use item dataset to get count of movies released in a particular year, store that value in a new column as 'total'). Now divide the value of each genre in that year by total to get percentage number of release in a particular year.
`(df.div(df['total'], axis= 0) * 100)`

Once that is achieved, you can either use univariate plots or can use the heatmap to visualise all the changes over the years 
in one go. 

Hint 3: Use groupby on the relevant column and use sum() on the same to find out the number of releases in a year/genre.

```python
#First cleaning up the dataframe

df = dfI.drop(['movie id','release date', 'movie title'], axis=1)
df['year'] = df['year'].astype('Int64') #making sure the years are the same without potential decimal issues

#Year as index, genre as columns

df2 = df.groupby('year').sum()


#Total number of movies in the year

df3 = df.groupby('year').agg(total_movie = pd.NamedAgg(column='year', aggfunc='count'))

#Merged into one usuable dataframe

merged_df = df3.merge(df2, on='year')

#Calculating the Genre % vs Released movies

df4 = (merged_df.div(merged_df['total_movie'], axis= 0) * 100).round(2)
df4 = df4.drop(['total_movie'], axis=1)

#Showing result as a heatmap
result = sns.heatmap(df4, cmap="YlGnBu")

df4
```

**Insights:**

Considering the years which only one movie were released as outliers. Drama has benefited most by an increase in popularity, it is also the most popular genre of movie in the data provided.

Conversley, documentary as a genre has remained low and not seen a benefit in popularity rise over time.

War is the genre that has decreased the most over time. However, the number of movies for older dates is low and therefore this should not be considerd accurate.



### 9. Find the top 25 movies in terms of average ratings for movies that have been rated more than 100 times - 10 marks

Hints : 

1. Find the count of ratings and average ratings for every movie.
2. Slice the movies which have ratings more than 100.
3. Sort values according to average rating such that movie which highest rating is on top.
4. Select top 25 movies.
5. You will have to use the .merge() function to get the movie titles.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
#Creating a DataFrame with just the movie id and title to be merged later

movie_names = dfI[['movie id', 'movie title']]

#Calculating the count and average (mean) of the ratings for each movie
count_df = dfD.groupby('movie id').agg(rating_count = pd.NamedAgg(column='movie id', aggfunc='count'))
average_df = dfD.groupby('movie id').agg(rating_avg = pd.NamedAgg(column='rating', aggfunc='mean'))

#Slicing only movies with rating counts over 100 (excluding 100 because instructions explicitly said "more than 100")
count_df = count_df[count_df['rating_count'] > 100]

#Merging the movies that have more than 100 ratings with the movie name dataframe
top_movies = count_df.merge(average_df, on='movie id')
top_movies = top_movies.merge(movie_names, on='movie id')

#Sorting the movies based on their rating (largest to smallest)
top_movies = top_movies.sort_values(by='rating_avg', ascending=False)
top_movies.set_index('movie id', inplace=True)

#Selecting only the top 25 movies.
top25 = top_movies[['movie title', 'rating_avg', 'rating_count']]
top25 = top_movies.nlargest(25, 'rating_avg')
top25.set_index('movie title', inplace=True)

result = top25

#Final Top 25 Movies. Their rating and the rating count
result
```

### 10. Check for the validity of the below statements with respect to the data provided - 10 marks

**compare the percentages**

3. you might want ot use the .sum(), .div() function here.

4. Use number of ratings to validate the numbers. For example, if out of 4000 ratings received by women, 3000 are for drama, we will assume that 75% of the women watch drama.

```python
#Merging all datasets
df = dfD.merge(dfU, on='user id')
df = df.merge(dfI, on='movie id')

#This is the total number of ratings provided by men and women.
df2 = df
df2 = df2['gender'].value_counts()


#Splitting the value according to gender
total_male = df2.values[0]
total_female = df2.values[1]

#Calculating the count of ratings for Drama based on gender
df1 = df[df['Drama'] >= 1]
count_g = df1[['gender', 'Drama']].groupby('gender').count()
#Assigning the gender totals to their own variable
no_female = count_g.loc['F']
no_male = count_g.loc['M']

#Calculating the percentage of men who watch Drama (number of men who rated drama / count of all ratings by men * 100)
perc_male = no_male / total_male * 100
print('Percentage of men who watch Drama:', perc_male.values[0].round(2))

#Calculating the percentage of women who watch Drama (number of women who rated drama / count of all ratings by men * 100)
perc_female = no_female / total_female * 100
print('Percentage of women who watch Drama:', perc_female.values[0].round(2))

#Based on the percentage of men/women who watch Drama, we validate the True/False statements. 

#I then do the same for other genres below.

print()
df1 = df[df['Sci-Fi'] >= 1]
count_g = df1[['gender', 'Sci-Fi']].groupby('gender').count()
no_female = count_g.loc['F']
no_male = count_g.loc['M']

perc_male = no_male / total_male * 100
print('Percentage of men who watch Sci-Fi', perc_male.values[0].round(2))

perc_female = no_female / total_female * 100
print('Percentage of women who watch Sci-Fi', perc_female.values[0].round(2))


print()
df1 = df[df['Romance'] >= 1]
count_g = df1[['gender', 'Romance']].groupby('gender').count()
no_female = count_g.loc['F']
no_male = count_g.loc['M']

perc_male = no_male / total_male * 100
print('Percentage of men who watch Romance:', perc_male.values[0].round(2))

perc_female = no_female / total_female * 100
print('Percentage of women who watch Romance:', perc_female.values[0].round(2))

```
